Process mouse homology sheet and add annotation transfers

```python
homology_table = pd.read_csv('./data/NHP_BG_joint_annotation-HMBA_Primate_Mouse_homology.tsv', sep='\t')
homology_table.head(5)
```

```python
from cell_annotation_schema.datamodel.bican.cell_annotation_schema import AnnotationTransfer
from cell_annotation_schema.taxonomy_utils import get_annotation_with_name

failed_names = list()
for index, row in homology_table.iterrows():
    # hack cluster name 10_NN -> NN_10
    mq_name_split = row["macaque_cluster"].split("_")
    mq_name = mq_name_split[1] + "_" + mq_name_split[0]

    bg_annotation = get_annotation_with_name(taxonomy, mq_name, "cluster")
    for mouse_label in row["homology_subclass"].split("|"):
        mouse_annotation = get_annotation_with_name(mouse_taxonomy, mouse_label, "subclass", compare_func=lambda x, y: x.endswith(y))
    
        if not bg_annotation:
            failed_names.append("Couldn't find BG annotation: " + row["macaque_cluster"])
        if not mouse_annotation:
            failed_names.append("Couldn't find Mouse annotation: " + row["homology_subclass"])
        
        if bg_annotation and mouse_annotation:
            at = AnnotationTransfer()
            at.transferred_cell_label = mouse_annotation.cell_label
            at.source_taxonomy = "https://purl.brain-bican.org/taxonomy/CCN20230722/"
            at.source_node_accession = mouse_annotation.cell_set_accession
            at.algorithm_name = "MapMyCell"
            bg_annotation.transferred_annotations = [at]
            print(bg_annotation.cell_label + "." + str(at))

for failure in failed_names:
    print(failure)
```

Add cell_ids to the taxonomy from anndata

```python
from cas.populate_cell_ids import add_cell_ids
from jsonasobj2 import JsonObj
from dataclasses import asdict

taxon = asdict(taxonomy)
add_cell_ids(taxon, ad, [lbl_set["name"] for lbl_set in cas_json["labelsets"]])

# cas-linkml & cas-tools integration workaround
for annotation in taxon["annotations"]:
    if annotation.get("author_annotation_fields") and isinstance(annotation["author_annotation_fields"], JsonObj):
        annotation["author_annotation_fields"] = annotation["author_annotation_fields"]._as_dict
        
taxon["annotations"][0]["cell_ids"][0:5]
```
